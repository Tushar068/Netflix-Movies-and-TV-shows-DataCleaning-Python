```python
#Importing the necessary libraries

import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading the dataset

df = pd.read_csv('Netflix_Movies_and_TV_Shows.csv')
df.head()
```

```python
df.info()
```

```python
# Rows and Columns

print("Number of Rows:", df.shape[0])
print("Number of Columns:", df.shape[1])
```

```python
# Checking for duplicates

print(df.duplicated().sum())
```

```python
# Check for missing values

df.isnull().sum()
```

```python
# Handling missing values
# For 'Director', 'Cast', 'Country', 'Rating' we can fill missing values with 'Unknown' 

df['director'] = df['director'].fillna('Unknown', inplace=True)
df['cast'] = df['cast'].fillna('Unknown', inplace=True)
df['country'] = df['country'].fillna('Unknown', inplace=True)
df['rating'] = df['rating'].fillna('Unknown', inplace=True)


print(df.isnull().sum())
```

```python
# Extracting the year from 'date_added' and creating a new column 'year_added'

df['year_added'] = df['date_added'].str.extract(r'(\d{4})')
df['year_added']
```

```python
df['year_added'].isnull().sum()
```

```python
# Fill nulls in year_added with mode

mode_year = df['year_added'].mode()[0]
df['year_added'] = df['year_added'].fillna(mode_year)
print(df['year_added'].isnull().sum())
```

```python
# Converting 'year_added' to integer type

df['year_added'] = df['year_added'].astype('Int64')
print(df['year_added'].head())
```

```python
# Fill nulls in date_added with mode

mode_date = df['date_added'].mode()[0]
df['date_added'] = df['date_added'].fillna(mode_date)
print(df['date_added'].isnull().sum())
```

```python
# Converting 'date_added' to datetime type

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
print(df['date_added'].dtype)
```

```python
# Removing leading and trailing spaces from string columns

df = df.apply(lambda x: x.str.strip() if x.dtype == 'object' else x)
```

```python
# Renaming the 'listed_in' column to 'genre'

df.rename(columns={'listed_in': 'genre'}, inplace=True)
print(df.columns)
```

```python
# Removing '#' from the 'title' column

df['title'] = df['title'].str.replace('#', '', regex=False)
```

```python
df.info()
```

```python
# Saving the cleaned dataset to a new CSV file
df.to_csv('Netflix_Cleaned.csv', index=False)
print("Saved!")
```
